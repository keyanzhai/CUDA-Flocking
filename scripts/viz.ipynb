```python
from matplotlib import pyplot as plt 
```

```python
# Data for numBods vs fps
numBoids = [5000, 10000, 20000, 50000, 100000, 200000, 500000, 1000000]
numBoids_fps_naive_viz_enabled = [445, 210, 106, 28, 8.6, 2.4, 0, 0]
numBoids_fps_naive_viz_disabled = [510, 225, 110, 28, 8.6, 2.4, 0, 0]
numBoids_fps_scattered_viz_enabled = [1180, 930, 690, 430, 150, 65, 10.8, 1]
numBoids_fps_scattered_viz_disabled = [1650, 1200, 870, 500, 155, 70, 10.8, 1]
numBoids_fps_coherent_viz_enabled = [1300, 1170, 1090, 925, 570, 310, 105, 38]
numBoids_fps_coherent_viz_disabled = [1980, 1730, 1560, 1200, 820, 440, 130, 40]
```

```python
plt.figure(facecolor='white')
plt.plot(numBoids, numBoids_fps_naive_viz_enabled, color="r") 
plt.plot(numBoids, numBoids_fps_scattered_viz_enabled, color="g") 
plt.plot(numBoids, numBoids_fps_coherent_viz_enabled, color="b") 
plt.title("FPS vs Number of boids\n(visualization enabled)")
plt.legend(["naive", "scattered uniform grid", "coherent uniform grid"], loc="upper right")
plt.xlabel("Number of boids")
plt.ylabel("FPS")
plt.savefig("../images/FPSvsBoids1.png") 
plt.show()
```

```python
plt.figure(facecolor='white')
plt.plot(numBoids, numBoids_fps_naive_viz_disabled, color="r", alpha=0.5) 
plt.plot(numBoids, numBoids_fps_scattered_viz_disabled, color="g", alpha=0.5) 
plt.plot(numBoids, numBoids_fps_coherent_viz_disabled, color="b", alpha=0.5) 
plt.title("FPS vs Number of boids\n(visualization disabled)")
plt.legend(["naive", "scattered uniform grid", "coherent uniform grid"], loc="upper right")
plt.xlabel("Number of boids")
plt.ylabel("FPS")
plt.savefig("../images/FPSvsBoids2.png") 
plt.show()
```
